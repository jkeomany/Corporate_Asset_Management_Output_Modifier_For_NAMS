```python
# Imports the required python packages

!pip install pandas
!pip install openpyxl
import pandas as pd
```

```python
# Reads the current excel file

file = pd.read_excel(file_name)
file
```

```python
# Creates a copy of your file

df = file
```

```python
# Removes the last 3 characters of the estimate (ie. ',00')

df['Estimate'] = df['Estimate'].str[:-3]
df
```

```python
# Creates the new Excel file

df.to_excel('NAMS_Editor.xlsx', index=False)
```
